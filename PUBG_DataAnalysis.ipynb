<h1>2. Best guns are having 7.62 bullets?

```python
a=WeaponStats_RemovedColumns[WeaponStats_RemovedColumns['Bullet Type']==7.62]
mean_damage_A= a[['Damage','Range','Bullet Speed','Rate of Fire']].mean()
total_A=mean_damage_A.sum()

b=WeaponStats_RemovedColumns[WeaponStats_RemovedColumns['Bullet Type']==5.56]
mean_damage_B= b[['Damage','Range','Bullet Speed','Rate of Fire']].mean()
total_B=mean_damage_B.sum()

c=WeaponStats_RemovedColumns[WeaponStats_RemovedColumns['Bullet Type']==9.00]
mean_damage_C= c[['Damage','Range','Bullet Speed','Rate of Fire']].mean()
total_C=mean_damage_C.sum()

d=WeaponStats_RemovedColumns[WeaponStats_RemovedColumns['Bullet Type']==0.45]
mean_damage_D= d[['Damage','Range','Bullet Speed','Rate of Fire']].mean()
total_D=mean_damage_D.sum()

e=WeaponStats_RemovedColumns[WeaponStats_RemovedColumns['Bullet Type']==12.00]
mean_damage_E= e[['Damage','Range','Bullet Speed','Rate of Fire']].mean()
total_E=mean_damage_E.sum()



total=pd.DataFrame({"total damage": [total_A,total_B,total_C,total_D,total_E]},index=[0,1,2,3,4])
total=total.sort_values('total damage',ascending=False)
largest_dam=total.iloc[0,:][0]


if(largest_dam==total_A):
    print("7.62 are the best Guns!!")    
elif(largest_dam==total_B):
    print("5.56 are the best Guns!!")
elif(largest_dam==total_C):
    print("9.00 are the best Guns!!")
elif(largest_dam==total_D):
    print("0.45 are the best Guns!!")
elif(largest_dam==total_E):
    print("12.00 are the best Guns!!")
else:
    print("AWM is Best!!!")
```

<h3 style="color: blue"> Hence Guns With Bullets <mark>5.56</mark> are the best Guns Considering DAMAGE,RANGE,RATE OF FIRE & BULLET SPEED

<h1>6. More kills insure winning the game?

```python
mean_kills=MatchData_removedColmuns['kills'].mean()
a=MatchData_removedColmuns[MatchData_removedColmuns['kills']<mean_kills]
less_than_avg=a.groupby('kills')['winPoints'].mean().max()
b=MatchData_removedColmuns[MatchData_removedColmuns['kills']>mean_kills]
more_than_avg=b.groupby('kills')['winPoints'].mean().max()

if(less_than_avg<more_than_avg):
    print("YES, more kills ensures winning the game")
else:
    print("NO, more killes do not have any relation towards winning the game")
```

<h3 style="color:blue">YES, more kills ensures winning the game

<h1>7. Number of heals and revives increase chance of winning?

```python
mean_recovered=MatchData_removedColmuns[['heals','revives']].mean()
a=pd.DataFrame({'heals':mean_recovered[0],'revives':mean_recovered[1]},index=['Mean'])
temp=MatchData_removedColmuns[['heals','revives','winPoints']]
temp['avg_heal']=mean_recovered[0]
temp['avg_revives']=mean_recovered[1]
less_than_avg=temp[(temp['heals'] <temp['avg_heal']) & (temp['revives']<temp['avg_revives'])]
more_than_avg=temp[(temp['heals'] >temp['avg_heal']) & (temp['revives']>temp['avg_revives'])]

sum_lta=less_than_avg.groupby(['heals'])['winPoints'].mean().sum()
sum_mta=more_than_avg.groupby(['heals'])['winPoints'].mean().sum()


if(sum_mta>sum_lta):
    print("YES,heals & revives increase the chance of winning!")
else:
    print("No, heals & revives have no relation in chance of winning!")
```

<h3 style="color:blue">YES,heals & revives increase the chance of winning!

<h1>8. Headshots are key to success?

```python
temp=MatchData_removedColmuns['headshotKills']
avghs=MatchData_removedColmuns['headshotKills'].mean()
less_than_avg=temp[(temp<avghs)].mean()
more_than_avg=temp[(temp>avghs)].mean()

if(more_than_avg>less_than_avg):
    print("YES, headshots are really key to success!!!")
else:
    print("NO, headshots are really key to success!!!")
    
```
